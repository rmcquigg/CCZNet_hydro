```python
import pandas as pd
import plotly.express as px
from google.colab import drive
drive.mount("/content/drive/")
```

```python
#@title Enter file paths to combine:
#@markdown Combine the two files into one table on the common timestamp index. This assumes the left-most column in each file is the date time field.

#@markdown This will only work if each file has distinct headers (i.e. if both files have the site name as a header, you will need to add a suffix to differentiate between them).

#@markdown ---
file1='/content/drive/Shared drives/CZ Collaborative/Data Management/Data/FILENAMEHERE.csv' #@param {type:"string"}
file2='/content/drive/Shared drives/CZ Collaborative/Data Management/Data/FILENAMEHERE.csv' #@param {type:"string"}
```

```python
#@title Select 3 columns to plot
import ipywidgets as widgets
df1=pd.read_csv(file1,index_col=0,na_values='NAN')
df2=pd.read_csv(file2,index_col=0,na_values='NAN')
data=df1.merge(df2,how='inner',left_index=True, right_index=True)

col1=widgets.Dropdown(options=list(data.columns.values),description='Column 1:')
col2=widgets.Dropdown(options=list(data.columns.values),description='Column 2:')
col3=widgets.Dropdown(options=list(data.columns.values),description='Column 3:')
display(col1,col2,col3)
```

##If you added more than the original 3 columns to plot, add them below in a similar format to the existing code (inside the brackets and separated by a comma).##

*To save, click the camera icon in the top right corner of the plot.*

*To invert the y-axis (like for looking at DTW, which is negative when above ground surface), uncomment (remove the #) from the beginning of the second line.*

```python
#@title Plot
fig=px.line(data,y=[col1.value,col2.value,col3.value])
#fig['layout']['yaxis']['autorange'] = "reversed"
fig.show()
```
